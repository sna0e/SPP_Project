# 학과별 처리

```python
school_list = ["linguistics_literature", "humanities", "business_economics",
               "law", "social_science", "education", "nursery_education", "special_education",
               "elementary_education", "secondary_education", "architecture", "civil_engineering",
               "electricity_electronics", "energy_engineering", "material_engineering", "computer_engineering", 
               "industrial_engineering", "chemical_engineering", "miscellaneous_engineering", "agriculture_science", "biostudies_science", 
               "human_ecology_science", "physics_science", "medicines", "nursing", "pharmacy", "health",
               "design_arts", "applied_arts", "dance_arts", "fine_arts", "theater_arts", "music_arts"]
col = ["year_list", "linguistics_literature", "humanities", "business_economics",
               "law", "social_science", "education", "nursery_education", "special_education",
               "elementary_education", "secondary_education", "architecture", "civil_engineering",
               "electricity_electronics", "energy_engineering", "material_engineering", "computer_engineering", 
               "industrial_engineering", "chemical_engineering", "miscellaneous_engineering", "agriculture_science", "biostudies_science", 
               "human_ecology_science", "physics_science", "medicines", "nursing", "pharmacy", "health",
               "design_arts", "applied_arts", "dance_arts", "fine_arts", "theater_arts", "music_arts"]
```

```python
competition_rate_1 = pd.DataFrame(columns=col)
competition_rate_2 = pd.DataFrame(columns=col)
```

```python
len(school_list)
```

```python
year_list = np.arange(2011, 2023)
year_list
```

```python
for s in school_list :
    temp = pd.read_csv(f"./학과데이터/"+s+".csv", index_col="year")
    temp = pd.DataFrame(temp.T)
    print(s)
    competition_rate_1[s] = temp.loc[:, "competition_rate_1"]
    competition_rate_2[s] = temp.loc[:, "competition_rate_2"]
```

```python
temp
```

```python
competition_rate_1["year_list"] = year_list
competition_rate_1
```

# computer engineering

```python
ce = pd.DataFrame()
ce["computer_engineering"] = competition_rate_1_merge["computer_engineering"]
ce["ai"] = competition_rate_1_merge["ai"]
ce["graphic_card"] = competition_rate_1_merge["graphic_card"]
ce["chatbot"] = competition_rate_1_merge["chatbot"]
ce["copyright"] = competition_rate_1_merge["copyright"]
ce["autonomous"] = competition_rate_1_merge["autonomous"]
ce["bigdata"] = competition_rate_1_merge["bigdata"]
ce["data_analytics"] = competition_rate_1_merge["data analytics"]
```

```python
ce
```

```python
from statsmodels.formula.api import ols, glm
from statsmodels.graphics.regressionplots import plot_partregress_grid

model = ols('computer_engineering~graphic_card+chatbot+copyright+autonomous+bigdata+data_analytics', data=ce).fit()
model.summary() # 각 독립변수의 영향력 정도 p < 0.05 여야 유의미함
```

```python
fig=plot_partregress_grid(model)
fig.tight_layout(pad=1)
```
